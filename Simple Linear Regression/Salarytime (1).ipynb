```python
#Linear Transformations

#X[['YearsExperience']] = X[['YearsExperience']]**2
X[['YearsExperience']] = np.sqrt(X[['YearsExperience']])
#X[['YearsExperience']]  = np.log(X[['YearsExperience']])
#X[['YearsExperience']] = 1 / np.sqrt(X[['YearsExperience']])
#X[['YearsExperience']] = 1 / np.cbrt(X[['YearsExperience']])
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X,Y)
Y_pred=LR.predict(X)
Y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y,Y_pred)
print("Mean Squared Error",mse.round(3))
```

```python
import numpy as np
Rmse=np.sqrt(mse)
print("root Mean Squared Error",Rmse.round(3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y,Y_pred)
(r2*100).round(3)
```
