```python
#Splitting Variables as X and  Y
X=df[["Sorting Time"]]
Y=df["Delivery Time"]
```

```python
#Linear Transformations

#X[['Sorting Time']] = X[['Sorting Time']]**2
X[['Sorting Time']] = np.sqrt(X[['Sorting Time']])
#X[['Sorting Time']]  = np.log(X[['Sorting Time']])
#X[['Sorting Time']] = 1 / np.sqrt(X[['Sorting Time']])
#X[['Sorting Time']] = 1 / np.cbrt(X[['Sorting Time']])
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X,Y)
Y_pred=LR.predict(X)
Y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y,Y_pred)
print("Mean Squared Error",mse.round(3))
```

```python
import numpy as np
Rmse=np.sqrt(mse)
print("root Mean Squared Error",Rmse.round(3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y,Y_pred)
(r2*100).round(3)
```
